```python
#实验步骤1:
#将一个一定范围的随机分布数据变换到另一个范围的随机分布数据
import random
import numpy as np
from matplotlib import pyplot as plt
f = plt.figure(figsize=(6,10)) #调整输出图片尺寸
ax1 = f.add_subplot(3,1,1)
ax1.plot([0,1],[1,3],color='b',linestyle='-') #画出y=2x+1的直线
i = 0
M = 100
x_list = []
y_list = []
for i in range(M):
    x = random.uniform(0,1)
    x_list.append(x)            
    y = 2*x + 1
    y_list.append(y)
    
    
ax2 = f.add_subplot(3,1,2)
ax2.scatter(x_list,np.zeros(len(x_list)),marker='o',color='r')
ax2.plot([0,0],color='b',linestyle='--')
ax3 = f.add_subplot(3,1,3)
ax3.scatter(y_list,np.zeros(len(y_list)),marker='o',color='r')
ax3.plot([0,3],[0,0],color='b',linestyle='--')

plt.show()
```

```python
#实验步骤2:
#将一个一定范围的均匀分布数据变换到另一个范围的正态分布数据
#我们定义的变换函数如下
import random
import math
from scipy.special import erfinv
from matplotlib import pyplot as plt
mu = 0
sigma = 1
i = 0
a = 0.001
a_b_array = [[0,0]]
for i in range(500):
    b = mu + math.sqrt(2.0)*sigma*erfinv(2.0*a - 1.0)
    a_b_array.append([a,b])
    plt.plot(a,b,'ro')
    i += i
    a = 0.01*i
plt.show()
```

```python
#运用以上定义的变换函数
import random
import math
from scipy.special import erfinv
from matplotlib import pyplot as plt
f = plt.figure(figsize=(6,10))
M = 1000
x_list = []
y_list = []
for i in range(M):
    x = random.uniform(0,1)
    x_list.append(x)
    mu = 0
    sigma = 1
    y = mu + math.sqrt(2.0)*sigma*erfinv(2.0*x - 1.0) 
    y_list.append(y)
ax1 = f.add_subplot(211)
ax1.hist(x_list,bins=10)
ax2 = f.add_subplot(212)
ax2.hist(y_list,bins=10)
plt.show()
```
