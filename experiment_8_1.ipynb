```python
#实验步骤1:
#生成 M=1 个随机点
import random #导入random包
from matplotlib import pyplot as plt #导入绘图工具
a = random.uniform(0,1) #在「0，1）的范围内随机生成点
plt.plot(a,0,marker='o',color='r') #画出点(a,0)
plt.plot([0,0],color='b',linestyle='--') #画出0到1的直线
plt.show() #展示结果
print (a)  #输出随机数a
```

```python
#实验步骤2:
#生成 M=100 个随机点
import random
from matplotlib import pyplot as plt
i = 0
M = 100
for i in range(M):
    b = random.uniform(0,1)
    plt.plot(b,0,marker='o',color='r')
    plt.plot([0,0],color='b',linestyle='--')
    i += 1
plt.show()
```

```python
#用直方图来刻画随机性
import random
from matplotlib import pyplot as plt
M = 100
b_array = [random.uniform(0,1) for i in range(M)]
plt.hist(b_array,bins=10)
plt.show()

#思考，如何使上面的散点图和直方图表示同一个随机分布
```
